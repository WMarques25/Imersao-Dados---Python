```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})

df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

df['remoto'].value_counts()
```

```python
import numpy as np

# DataFrame de testes
df_salarios = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "salario": [4000, np.nan, 5000, np.nan, 100000]
})

# Calcula a media e substitui os valores nulos arredondando
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))

# Calcula a mediana e substitui os valores nulos arredondando
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios["salario"].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})

# preenchendo com FowardFill, completa com o anterior.
df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].fillna(method="ffill")

# preenchendo com BackFill, completa com o posterior.
df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].fillna(method="bfill")

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade": ["São Paulo", np.nan, "Curitiba", np.nan, "São Paulo"]
})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não informado")

df_cidades
```

```python
import plotly.express as px
```

```python
fig = px.bar(
    df_limpo.groupby("senioridade")["usd"].mean().reset_index(),
    x="senioridade",
    y="usd",
    title="Salário Médio por Senioridade (Plotly)",
    labels={"senioridade": "Senioridade", "usd": "Salário Médio (USD)"},
    color="senioridade"
)
fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray': ordem_senioridade})
fig.show()
```

```python
remoto_contagem = df_limpo["remoto"].value_counts().reset_index()
remoto_contagem.columns = ["tipo_trabalho", "quantidade"]

fig = px.pie(
    remoto_contagem,
    names="tipo_trabalho",
    values="quantidade",
    title="Proporção dos Tipos de Trabalho",
    hole=0.4
)
fig.update_traces(textinfo="percent+label")
fig.show()
```

```python
df_dataScientist = df_limpo[df_limpo["cargo"] == "Data Scientist"].reset_index()

# df_dataScientist.info()
# df_dataScientist.head()
pais_salario = df_dataScientist.groupby("empresa")["usd"].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    pais_salario,
    x="empresa",
    y="usd",
    title="Salário Médio por País",
    labels={"empresa": "País", "usd": "Salário Médio (USD)"},
    color="empresa",
)
fig.show()
```

```python
import pycountry

# Converter ISO-2 -> ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criando coluna de ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Média Salarial Data Scientist por pais
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(
    media_ds_pais,
    locations='residencia_iso3',
    color='usd',
    title='Salário médio de Cientista de Dados por pais',
    color_continuous_scale='rdylgn',
    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'}
)
fig.show()
```
